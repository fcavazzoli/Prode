```python
def dates_format(date):
    date_new = date.split('-')
    date_new = int(date_new[0])
    return date_new
```

```python
tuples = [tuple(x) for x in history.values]
for x in tuples:
    x = tuple(x)
#history.reset_index(inplace=True)
```

```python
global match_list
match_list = []
def match_unique(x):
    global match_list
    match = (x[1],x[2])
    reverse_match = (x[2],x[1])
    if match in match_list:
        return x
    if reverse_match in match_list:
        return (x[0],x[2],x[1],x[4],x[3],x[5],x[6],x[7],x[8])
    match_list.append(match)
    return x
```

```python
def result(x1,x2):
    if x1>x2:
        return 0
    elif x1<x2:
        return 2
    return 1
```

```python
new = []
for x in tuples:
    x = match_unique(x)
    new.append(x)
```

```python
history = pd.DataFrame.from_records(new)
history = history[[0,1,2,3,4,5,7]]
history.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(history['home_team'])
history['home_team'] = le.transform(history['home_team'])
history['home_team'] = history.apply(lambda x: int(x['home_team']),axis=1)
le.fit(history['away_team'])
history['away_team'] = le.transform(history['away_team'])
history['away_team'] = history.apply(lambda x: int(x['away_team']),axis=1)
le.fit(history['tournament'])
history['tournament'] = le.transform(history['tournament'])
history['tournament'] = history.apply(lambda x: int(x['tournament']),axis=1)
le.fit(history['Place'])
history['Place'] = le.transform(history['Place'])
history['Place'] = history.apply(lambda x: int(x['Place']),axis=1)

history.fillna(0, inplace=True)
```

```python
def calcular_exactitud(resultados,prediccion):
    contador = 0
    for x in range (len(resultados)):
        if resultados[x] == prediccion[x]:
            contador += 1
    final = contador/len(resultados)*100
    print(final,"%")
```
